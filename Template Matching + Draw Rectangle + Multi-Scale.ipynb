### Template Matching + Draw Rectangle 
!!! TODO: Scaling !!!

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
template = cv2.imread('./Data/Reference/4.jpg') #1 is colour, 0 is greyscale, -1 is unchanged
template = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)

img_colour = cv2.imread('./Data/Test/2.jpg')
img = cv2.cvtColor(img_colour, cv2.COLOR_BGR2GRAY)
img_copy = img.copy()

if (img.size >= template.size):
    print 'ok'
    
else: 
    print "img.size is smaller than template.size"
    


w, h = template.shape[::-1]

print w, h
result = cv2.matchTemplate(img, template, cv2.TM_CCOEFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

top_left = max_loc
print top_left
bottom_right = (top_left[0] + int(w), top_left[1] + int(h))
print bottom_right

img_orig = cv2.imread('./Data/Test/2.jpg', -1)
cv2.rectangle(img_orig,top_left, bottom_right, (255,0,0), 20)

plt.subplot(221)
plt.imshow(template, cmap='gray')

plt.subplot(222)
plt.imshow(img_copy, cmap='gray')

plt.subplot(223)
plt.imshow(result, cmap = 'gray')

plt.subplot(224)
plt.imshow(cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB))

plt.show()
```
